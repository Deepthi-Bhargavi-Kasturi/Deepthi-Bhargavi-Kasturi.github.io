**What is a pipeline?**

This is an important API in the Transformers Library.







```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
classifier("I've been trying to avoid last minute plans my whole life.")
```

```python
classifier(
    ["I've been waiting for a HuggingFace course my whole life.", "I hate this so much!"]
)
```

```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification")
classifier(
    "This is a course about the Transformers library",
    candidate_labels=["education", "politics", "business"],
)

classifier(
    "This Anime is about Adventure and not giving up",
    candidate_labels=["Motivation","Anime","vacation"]
)

classifier(
    ["This Show is about Adventure and not giving up","This is a course about the Transformers library"],
    candidate_labels=["Inspiration","Anime","vacation"]
)
```
